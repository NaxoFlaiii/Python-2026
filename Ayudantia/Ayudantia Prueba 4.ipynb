```python
import matplotlib.pyplot as plt

lenguajes = ["Python", "C++", "Java", "JavaScript", "C#"]
preferencia = [45, 20, 15, 12, 8]


plt.bar(lenguajes, preferencia,)

plt.title("Preferencia de Lenguajes de Programación")
plt.xlabel("Lenguajes")
plt.ylabel("Porcentaje de preferencia")

plt.show()
```
