```python
datosdrop=datosduplicados.dropna().copy()
datosdrop
```

```python
datosdropsin=datosdrop.shape
print(f'La Cantidad de filas y columnas es: {datosdropsin}')
```

***Sección D***

```python
def filtro_fuego(Type1):
    return 'Fire' in Type1


solo_fuego = datosdrop[(datosdrop['Type 1'] == 'Fire') | (datosdrop['Type 2'] == 'Fire')]

solo_fuego.head(20)
```

```python
promedio_defensas = solo_fuego['Def'].mean()
print(f'El promedio De las defensas de los pokemons tipo fuego es de aproximadamente: {promedio_defensas:.2f}')
```

```python
nombres = solo_fuego['Name'].head(10)
poder_de_ataque = solo_fuego['Att'].head(10) # ¡Aquí agregamos el 10!


plt.bar(nombres, poder_de_ataque, color="blue")

plt.xticks(rotation=45, ha='right')


plt.show()
```
