```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

***Sección A***

```python
datos = pd.read_csv('../jupyter/data/Pokemon.csv', index_col = 'Number')
datos
```

```python
datosresumidos = datos.info()
datosresumidos
```

```python
datosdescritos = datos.describe()
datosdescritos
```

***Sección B***

```python
datos3 = datos.shape
print(f'La Cantidad de filas y columnas es: {datos3}')
```

***Sección C***

```python
datosduplicados = datos.drop_duplicates()
datosduplicados
```

```python
datosdrop = datosduplicados.dropna().copy()
datosdrop
```

```python
datosdropsin = datosdrop.shape
print(f'La Cantidad de filas y columnas es: {datosdropsin}')
```

***Sección D***

```python
def filtro_fuego(Type1):
    return 'Fire' in Type1


solo_fuego = datosdrop[(datosdrop['Type 1'] == 'Fire') | (datosdrop['Type 2'] == 'Fire')]

solo_fuego.head(20)
```

```python
promedio_defensas = solo_fuego['Def'].mean()
print(f'El promedio De las defensas de los pokemons tipo fuego es de aproximadamente: {promedio_defensas:.2f}')
```

**Sección F**

```python
nombres = solo_fuego['Name'].head(10)
poder_de_ataque = solo_fuego['Att'].head(10)

plt.Figure(figsize= (8 , 4))
plt.bar(nombres, poder_de_ataque, color = 'red')
plt.title('Top 10 de Pokemones segun su poder de ataque')
plt.xticks(rotation=45, ha='right')
for i, rating in enumerate(solo_fuego['Att']):
    plt.annotate(str(rating), xy=(i, rating), ha='center', va='bottom')

plt.show()
```
