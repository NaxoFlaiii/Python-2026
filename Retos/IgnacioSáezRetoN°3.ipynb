```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def preparardatos():
    horas = np.arange(1,11)
    porcentajes_cpu =[45, 58, 52, 70, 85, 92, 78, 64, 61, 50]
    
    data ={
        'Hora': horas,
        '% Uso CPU': porcentajes_cpu
    }

   
    df = pd.DataFrame(data)
    return df
```

```python
def  graficar_rendimiento(dataframe):
    fig, ax = plt.subplots() 

    ax.plot(dataframe['Hora'], dataframe['% Uso CPU'], marker='o', color='blue', linestyle='-')




    
    ax.set_title('Uso de CPU por horas')
    ax.set_xlabel('Horas de Simulación ')
    ax.set_ylabel('% Uso CPU')

    ax.grid(True)
    fig.savefig('grafico01',)
```

```python
df_usodecpuporhoras = preparardatos()# invocando la primera funcion 
graficar_rendimiento(df_usodecpuporhoras) # invocamdo segunda funcion con el primer parametro de la funcion
plt.show()
```
